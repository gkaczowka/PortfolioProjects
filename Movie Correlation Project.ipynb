```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv(r'C:\Users\gkacz\Desktop\gk\Portfolio\archive\movies.csv')

df.head()

#Ile jest braków danych

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Wyrzucenie wierszy z brakami danych

df = df.dropna(how='any',axis=0) 
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Typ danych w kolumnach
df.dtypes
```

```python
df['budget'] = df['budget'].astype('int64')
```

```python
df['gross'] = df['gross'].astype('int64')
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Wyrzucenie duplikatów

df.drop_duplicates()
```

```python

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross')
plt.xlabel('Gross')
plt.ylabel('Budget')
```

```python
# Budget vs gross w seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Korelacja r-pearsona

df.corr(method='pearson')
```

```python

# Duża korelacja pomiędzy budget i gross

correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.show()
```

```python
# Zmiana tekstowych kolumn (company, country itp.) na numeryczne odpowiedniki

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Korelacja dla wszystkich danych')

plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr = high_corr[(high_corr) != 1]
high_corr
```
